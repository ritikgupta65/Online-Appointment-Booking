```python
import numpy as np
import pandas as pd
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.layers import *
from tensorflow.keras import models
from tensorflow.keras.models import Sequential
from tensorflow.keras.regularizers import l2
```

```python
df = pd.read_csv(r"C:\Users\gupta\Downloads\human_vital_signs_dataset_2024.csv\human_vital_signs_dataset_2024.csv")
df
```

```python
df['Risk Category'] = df['Risk Category'].replace({
   'High Risk':1,
   'Low Risk':0 
})
new_df = df[['Heart Rate', 'Respiratory Rate', 'Systolic Blood Pressure', 'Diastolic Blood Pressure', 'Body Temperature', 'Oxygen Saturation' , 'Derived_Pulse_Pressure' ,'Risk Category']]
x = new_df.iloc[:, :-1]
y = new_df.iloc[:, -1]
y.value_counts()
```

```python
new_df
```

```python
tf.test.is_built_with_cuda()
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

model = Sequential()

model.add(Dense(64, input_dim=X_train.shape[1], activation='relu'))

model.add(Dense(32, activation='relu'))
model.add(Dense(16, activation='relu'))

model.add(Dense(1, activation='sigmoid')) 

model.compile(optimizer=Adam(), loss='binary_crossentropy', metrics=['accuracy'])

early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

history = model.fit(x ,y, epochs=10, batch_size=64, 
                    validation_split=0.2, callbacks=[early_stopping])

test_loss, test_acc = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {test_acc:.4f}")
```

```python
new_sample = np.array([[60	,12,	124,	86,	36.861707	,95.702046	,38]]) 
new_sample = scaler.transform(new_sample)  
prediction = model.predict(new_sample)
print(prediction)
print(f"Predicted Risk Category: {'High Risk' if prediction > 0.5 else 'Low Risk'}")
```

```python
# how to convert the xlsx file to txt file
import pandas as pd

# Load the Excel file
excel_file_path = r"C:\Users\gupta\Downloads\health_risk_dataset.xlsx"  # Replace with your file path
excel_data = pd.read_excel(excel_file_path, sheet_name=None)  # Load all sheets

# Prepare a combined text file
output_txt_path = "converted_file.txt"  # Specify the output .txt file name

with open(output_txt_path, "w") as txt_file:
    for sheet_name, data in excel_data.items():
        # Write the sheet name as a header
        txt_file.write(f"Sheet: {sheet_name}\n")
        txt_file.write(data.to_string(index=False))  # Write the sheet content
        txt_file.write("\n\n")  # Add spacing between sheets

print(f"Excel file successfully converted to: {output_txt_path}")

```
